```python
clients = pd.read_csv(Path.cwd() / 'data' / 'Clientes.csv')
```

```python
clients.head()
```

```python
first_names = ['Luana', 'Paulo', 'Romário', 'Victoria', 'Clarissa', 'Almeida', 'José']

clients.query(' first_name in @first_names ')
```

```python
clients.created_at = pd.to_datetime(clients.created_at)
```

```python
clients.dtypes
```

```python
clients['year'] = clients.created_at.dt.year
```

```python
clients
```

```python
clients.query(' year == 2017 ')
```

```python
clients.query(' year == 2017 and first_name in @first_names ')
```

```python
import datetime


# ano atual

current_year = datetime.datetime.now().year

clients.query(' year == @current_year ')
```
